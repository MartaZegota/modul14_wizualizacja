Zaimportowanie bazy danych

```python
import pandas as pd
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

Podział na zmienną objaśnianą y i zmienne objaśniające X

```python
features = ['Pregnancies','PlasmaGlucose','DiastolicBloodPressure','TricepsThickness','SerumInsulin','BMI','DiabetesPedigree','Age']
target = 'Diabetic'
X, y = diabetes[features], diabetes[target]
display(X)
display(y)
```

# Model pierwotny

Podzielenie danych na treningowe oraz testowe dla dwóch wybranych cech x

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X[['Pregnancies', 'Age']], y, test_size=0.30, random_state=0, stratify=y)
print ('Treningowe obserwacje: %d\nTestowe obserwacje: %d' % (X_train.shape[0], X_test.shape[0]))
```

Klasyfikator regresji logistycznej bez zdefiniowanych hiperparametrów

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler_2var = StandardScaler()
X_train_standardized = scaler_2var.fit_transform(X_train)
X_test_standardized = scaler_2var.fit_transform(X_test)
model_lr = LogisticRegression()
model_lr.fit(X_train_standardized, y_train)
```

Wizualizacja działania klasyfikatora w wersji pierwotnej

```python
import matplotlib.pyplot as plt
from numpy import where
from numpy import meshgrid
from numpy import arange
from numpy import hstack

def plot_classification_surface(X_plot, y_plot, trained_model):
    plt.figure(figsize=(12, 7))

    # określenie granic zbioru
    min1, max1 = X_plot[:, 0].min()-1, X_plot[:, 0].max()+1
    min2, max2 = X_plot[:, 1].min()-1, X_plot[:, 1].max()+1

    # skalowanie dla obu osi
    x1grid = arange(min1, max1, 0.1)
    x2grid = arange(min2, max2, 0.1)

    # utworzenie siatki
    xx, yy = meshgrid(x1grid, x2grid)

    # przetworzenie siatki w wektor
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))

    # stworzenie zmiennych objaśniających dla modelu
    grid = hstack((r1,r2))

    # predykcja modelu zbioru
    yhat = trained_model.predict(grid)

    # przetworzenie predykcji na decyzji dla konkretnego punktu wykresu
    zz = yhat.reshape(xx.shape)

    # wizualizacja siatki z przyporządkowaną klasą
    plt.contourf(xx, yy, zz, cmap='Paired')

    # tworzenie wykresu punktowego dla klas ze zbioru X
    for class_value in range(2):

        # przyporządkowanie klasy (y) do obsewacji (X)
        row_ix = where(y_plot == class_value)

        # stworzenie wykresu punktowego
        plt.scatter(X_plot[row_ix, 0], X_plot[row_ix, 1], cmap='Paired', alpha=0.3, label=class_value)

    # wizualizacja wykresu
    plt.legend(loc='upper right')
    plt.show()

plot_classification_surface(X_plot=X_train_standardized, y_plot=y_train, trained_model=model_lr)
```

```python
predictions_lr = model_lr.predict(X_test_standardized)
predictions_lr1 = model_lr.predict(X_train_standardized)
```

Porównanie prognozy z wartościami prawdziwymi za pomocą macierzy omyłek

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, predictions_lr)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix - Test\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_train, predictions_lr1)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix - Train\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

Metryki modelu pierwotnego: dokładność, precyzja, czułość, F1 score

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Metryki danych testowych")
print('Accuracy: ', accuracy_score(y_test, predictions_lr))
print("Precision:", precision_score(y_test, predictions_lr))
print("Recall:", recall_score(y_test, predictions_lr))
print("F1_score:", f1_score(y_test, predictions_lr))
print("\nMetryki danych treningowych")
print('Accuracy: ', accuracy_score(y_train, predictions_lr1))
print("Precision:", precision_score(y_train, predictions_lr1))
print("Recall:", recall_score(y_train, predictions_lr1))
print("F1_score:", f1_score(y_train, predictions_lr1))
```

# Model analizowany

Klasyfikator regresji logistycznej ze zdefiniowanymi hiperparametrami

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler_2var = StandardScaler()
X_train_standardized = scaler_2var.fit_transform(X_train)
X_test_standardized = scaler_2var.fit_transform(X_test)
model1_lr = LogisticRegression(penalty='l2', C=0.0001) # inicjalizacja modelu; podane hiperparametry
model1_lr.fit(X_train_standardized, y_train) # trenowanie modelu
```

Wizualizacja działania klasyfikatora ze zdefiniowanymi hiperparametrami

```python
from numpy import where
from numpy import meshgrid
from numpy import arange
from numpy import hstack

def plot_classification_surface(X_plot, y_plot, trained_model):
    plt.figure(figsize=(12, 7))

    # określenie granic zbioru
    min1, max1 = X_plot[:, 0].min()-1, X_plot[:, 0].max()+1
    min2, max2 = X_plot[:, 1].min()-1, X_plot[:, 1].max()+1

    # skalowanie dla obu osi
    x1grid = arange(min1, max1, 0.1)
    x2grid = arange(min2, max2, 0.1)

    # utworzenie siatki
    xx, yy = meshgrid(x1grid, x2grid)

    # przetworzenie siatki w wektor
    r1, r2 = xx.flatten(), yy.flatten()
    r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))

    # stworzenie zmiennych objaśniających dla modelu
    grid = hstack((r1,r2))

    # predykcja modelu zbioru
    yhat = trained_model.predict(grid)

    # przetworzenie predykcji na decyzji dla konkretnego punktu wykresu
    zz = yhat.reshape(xx.shape)

    # wizualizacja siatki z przyporządkowaną klasą
    plt.contourf(xx, yy, zz, cmap='Paired')

    # tworzenie wykresu punktowego dla klas ze zbioru X
    for class_value in range(2):

        # przyporządkowanie klasy (y) do obsewacji (X)
        row_ix = where(y_plot == class_value)

        # stworzenie wykresu punktowego
        plt.scatter(X_plot[row_ix, 0], X_plot[row_ix, 1], cmap='Paired', alpha=0.3, label=class_value)

    # wizualizacja wykresu
    plt.legend(loc='upper right')
    plt.show()

plot_classification_surface(X_plot=X_train_standardized, y_plot=y_train, trained_model=model1_lr)
```

W porównaniu do modelu pierwotnego, powyżej wprowadzono dwa hiperparametry klasyfikatora regresji logistycznej:

1) parametr C - czym mniejsza wartość (mniejsze wartości określają silniejszą regularyzację), tym większe zmiany na powyższej wizualizacji modelu; natomiast czym wartość C jest większa, tym mniejsze odchylenie wartości metryk modelu w odniesieniu do modelu pierwotnego, dla którego przyjęto wartość domyślną tego parametru, czyli 1

2) parametr penalty - wybrano l2, który wymusza, aby zmniejszyć wagi mniej znaczacych cech, ale ich nie zeruje; L2 daje lepsze przewidywanie, gdy zmienna wyjściowa jest funkcją wszystkich cech wejściowych

```python
predictions_lr2 = model1_lr.predict(X_test_standardized)
predictions_lr3 = model1_lr.predict(X_train_standardized)
```

Porównanie prognozy z wartościami prawdziwymi za pomocą macierzy omyłek - dane treningowe

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_train, predictions_lr3)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix - Train\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

Porównanie prognozy z wartościami prawdziwymi za pomocą macierzy omyłek - dane testowe

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, predictions_lr2)
ax = sns.heatmap(cm, annot=True, cmap='Blues', fmt='.0f')
ax.set_title('Confusion Matrix - Test\n\n')
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ')
plt.show()
```

Metryki modelu analizowanego: dokładność, precyzja, czułość, F1 score

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Metryki danych testowych")
print('Accuracy: ', accuracy_score(y_test, predictions_lr2))
print("Precision:", precision_score(y_test, predictions_lr2))
print("Recall:", recall_score(y_test, predictions_lr2))
print("F1_score:", f1_score(y_test, predictions_lr2))
print("\nMetryki danych treningowych")
print('Accuracy: ', accuracy_score(y_train, predictions_lr3))
print("Precision:", precision_score(y_train, predictions_lr3))
print("Recall:", recall_score(y_train, predictions_lr3))
print("F1_score:", f1_score(y_train, predictions_lr3))
```

Accuracy: Wartość metryki dokładność spadła w stosunku do modelu pierwotnego. Jednocześnie zbiór stał się niezbalansowany, co oznacza, że zaczęła dominować jedna z klas, w tym przypadku 0. Widać to na powyższej wizualizacji.

Precision: Natomiast wartość metryki precyzji wzrosła w stosunku do modelu pierwotnego, a co więcej w powyżej analizowanym modelu jej wartość wynosi aż 0,9 dla zbioru testowego, co świadczy o overfittingu.

Recall: Wartość metryki czułość znacząco spadła w porównaniu do tej zmierzonej dla modelu pierwotnego.

F1 score: Tym samym metryka F1, będąca średnią harmoniczną precyzyji i czułości, drastycznie zmalała z 0.5671752884257536 dla danych testowych modelu pierwotnego do 0.011920529801324504 dla modelu powyższego.

Z powyższych metryk wynika, że model działa gorzej niż jego pierwotna wersja przy przyjętych hiperparametrach klasyfikatora regresji logistycznej.
